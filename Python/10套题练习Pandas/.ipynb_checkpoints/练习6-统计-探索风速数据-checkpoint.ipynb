```python
#步骤1 导入必要的库
import pandas as pd
import datetime
```

```python
#步骤2 从以下地址导入数据
path = './datas/wind.data'
```

```python
#步骤3 将数据作存储并且设置前三列为合适的索引
data = pd.read_table(path, sep='\s+', parse_dates=[[0,1,2]])
```

```python
data #将原来的前3个分开的时间数据，合成1列，为Yr_Mo_Dy
```

```python
#步骤4 2061年？我们真的有这一年的数据？创建一个函数并用它去修复这个bug
```

```python
#定义一个修复函数
def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.datetime(year, x.month, x.day) #主要是修复年份，进行返回
```

```python
#在列上应用函数，并进行数据替换
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century) #对某一列上apply自定义函数，然后回传替换
```

```python
data.head() #可以看出，数据已修复
```

```python
#步骤5 将日期设为索引，注意数据类型，应该是datetime64[ns]
data.info() #Yr_Mo_Dy  datetime64
```

```python
#本来就是datetime类型，防止意外，再进行转换
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
```

```python
data = data.set_index('Yr_Mo_Dy') #传入column的名称，然后赋值data，为新的、修改过index的data
```

```python
#步骤11 对于每一个location，计算一月份的平均风速
```

```python
#创建一个新的一列，用于转换年、月、日
data['date'] = data.index #其实就是复制了INDEX
```

```python
data
```

```python
data['month'] = data['date'].apply(lambda x: x.month)
#添加了一个month列，该列提取了月份
#将date列，应用提取函数，函数是个匿名函数，传入每一条数据，返回月份
data['year'] = data['date'].apply(lambda x: x.year)
data['day'] = data['date'].apply(lambda x: x.day)
```

```python
data
```

```python
#以上，依据日期列，拆分出了三列，分别是月份列、年列、日子列
#思路是，复制index到某一列，再依次添加三列数据，基于日期列apply一个匿名函数即可
#lambda x: x.year() #注意，x为datetime数据
```

```python
january_winds = data.query('month == 1')
#query是查询的意思
#DataFrame调用query函数，传入一个判断条件，此处是'month == 1'
```

```python
january_winds
```

```python
january_winds.loc[:,'RPT':'MAL']
#采用标签索引，行我都要，列为左闭右闭，取除日期外所有列
```

```python
january_winds.loc[:,'RPT':'MAL'].mean()
#调用聚合函数，形成SERIRES,INDEX为各列，VALUE为各列的平均值
```

```python
#步骤12 对于数据记录按照年为频率取样
data.query('month == 1 and day == 1')
#采样效果，使用的是DATAFRAME的query函数，注意传递的参数
#'X == 1 and Y == 1'
```
