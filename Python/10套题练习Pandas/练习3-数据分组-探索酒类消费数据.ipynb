```python
#步骤1 导入必要的库
import pandas as pd
```

```python
#步骤2 从以下地址导入数据
path = './datas/drinks.csv'
```

```python
#步骤3 将数据框命名为drinks
drinks = pd.read_csv(path)
drinks.head()
```

```python
#步骤4 哪个大陆(continent)平均消耗的啤酒(beer)更多？
drinks.groupby(by = 'continent')['beer_servings'].mean()
#分组、选列、聚合
```

```python
#步骤5 打印出每个大陆(continent)的红酒消耗(wine_servings)的描述性统计值
drinks.groupby(by = 'continent')['wine_servings'].describe()
#聚合后，对GROUPBY对象调用describe方法，即可查看描述性统计值
```

```python
#步骤6 打印出每个大陆每种酒类别的消耗平均值
drinks.groupby('continent').mean()
#对continent分组后，对每个列都求平均值
```

```python
#步骤7 打印出每个大陆每种酒类别的消耗中位数
drinks.groupby('continent').median()
#GROUPBY调用median方法，即是求中位数
```

```python
#步骤8 打印出每个大陆对spirit饮品消耗的平均值，最大值和最小值
drinks.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min'])
```

```python
# 分组、选列、应用agg函数进行聚合（要想实施单个聚合函数，直接.max（），要想多个，往agg函数中传递列表，列表包含多个聚合函数）
```
